```python
# TARGET GUIDED ORDINAL ENCODING
# 1.ordering the labels according to the target
# 2.replace the labels by the joint probability of being 1 or 0
```

```python
a=pd.read_csv("train(2).csv",usecols=['Cabin','Survived'])
```

```python
a
```

```python
a.Cabin.fillna('Missing',inplace=True)
```

```python
a
```

```python
a['Cabin'].astype('str')
```

```python
a['Cabin']=a['Cabin'].str[0]
```

```python
a.head(5)
```

```python
a.Cabin.unique()
```

```python
a.groupby(['Cabin'])['Survived'].mean()
```

```python
a.groupby(['Cabin'])['Survived'].value_counts()
```

```python
a.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
ordinal_labels=a.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
# enumerate=this will rank the data based on the positions like T=0,M=1 etc
ordinal_labels2={k:i for i,k in enumerate(ordinal_labels,1)}
ordinal_labels2
```

```python
a['Cabin_rank']=a['Cabin'].map({'T': 1, 'M': 2, 'A': 3, 'G': 4, 'C': 5, 'F': 6, 'B': 7, 'E': 8, 'D': 9})
```

```python
a
```

```python
# MEAN ENCODING(replace the categorical variables with mean)
mean_coding=a.groupby(['Cabin'])['Survived'].mean().to_dict()
```

```python
a['Mean_encoding']=a['Cabin'].map(mean_coding)
```
