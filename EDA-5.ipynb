```python
import pandas as pd 
import numpy as np
```

```python
# ####PROBABILITY RATIO ENCODING 
# 1.PROB OF SURVIVED BASED ON CABIN 
# 2.PROB OF DEAD......1-PROB_SURVIVED 
# 3.RATIO=LIVED/DEAD
# 4.TO DICT AND MAP IN DF
# 5.REPLACE WITH CATEGORICAL FEATURE 
```

```python
df=pd.read_csv('train(2).csv',usecols=['Cabin','Survived'])
```

```python
df['Cabin'].fillna('Missing',inplace=True)
```

```python
df['Cabin'].astype('str')
```

```python
df['Cabin']=df['Cabin'].str[0]
```

```python
df.head(10)
```

```python
df.Cabin.unique()
```

```python
# probability ratio encoding
prob_df=df.groupby(['Cabin'])['Survived'].mean()
```

```python
prob_df=pd.DataFrame(prob_df)
```

```python
prob_df
```

```python
# prob of died and survived == 1
prob_df['Died']=1-prob_df['Survived']
```

```python
prob_df
```

```python
prob_df['Probability_ratio']=prob_df['Survived']/prob_df['Died']
```

```python
prob_df
```

```python
prob_encoded=prob_df['Probability_ratio'].to_dict()
```

```python
df['Cabin_encoding']=df['Cabin'].map(prob_encoded)
```
