```python
# based on the ranks we categorise the data 
# example==grading system(a=1,b=2,c=3,d=4,e=5,f=6)
# experience of a batsman(10y=1,5y=2,3y=3,1y=4)
# the weekdays and weekends (m,t,w,t,f,s,s) we can give good ranks by customer information
import datetime
import pandas as pd 
import numpy as np
```

```python
date_time=datetime.datetime.today()
date_time
```

```python
data=[date_time-datetime.timedelta(x) for x in range(0,15)]
data
```

```python
data1=pd.DataFrame(data)
data1.columns=['days']
```

```python
data1
```

```python
data1['Hour']=data1['days'].dt.hour
```

```python
data1['Minutes']=data1['days'].dt.minute
```

```python
data1['Seconds']=data1['days'].dt.second
```

```python
data1['Micro_Seconds']=data1['days'].dt.microsecond
```

```python
data1['weekday']=data1['days'].dt.weekday
```

```python
data1
```

```python
data1['weekdayname']=data1['weekday'].map({0:'Monday',1:'Tuesday',2:'Wednesday',3:'Thursday',4:'Friday',5:'Saturday',6:'Sunday'})
```
